```python
# KOREAN NOVEL GLOSSARY MAKER - FIXED VERSION

# REPLACE your old API key input code with this:
from google.colab import userdata
import os

# Securely get the API key from Colab Secrets
try:
    GEMINI_API_KEY = userdata.get('GEMINI_API_KEY')
    # Also set it as an environment variable for the Gemini library
    os.environ["GEMINI_API_KEY"] = GEMINI_API_KEY
    print("‚úì API key loaded from Colab Secrets.")
except userdata.SecretNotFoundError:
    print("‚ö†Ô∏è Secret 'GEMINI_API_KEY' not found. Please set it in the Colab Secrets panel (left sidebar key icon).")
    # Optionally, you can fall back to input() or raise an error
    raise

# Setup
!git clone https://github.com/foodiemonster007/Korean-Novel-Glossary-Maker
%cd Korean-Novel-Glossary-Maker
!pip install -r requirements.txt

# Create proper config FIRST (before uploading files)
import json
config = {
    "API_KEY": GEMINI_API_KEY,
    "MODEL_NAME": "gemini-2.5-flash",
    "RAWS_FOLDER": "raws",
    "NOUNS_JSON_FILE": "nouns.json",
    "REFERENCE_FILE": "",
    "OUTPUT_EXCEL": "glossary.xlsx",
    "CHAPTERS_ANALYZED": 10,
    "CATEGORIZATION_BATCH_SIZE": 50,
    "TRANSLATION_BATCH_SIZE": 50,
    "HANJA_GUESSING_BATCH_SIZE": 50,
    "MAX_RETRIES": 3,
    "RETRY_DELAY": 31,
    "HANJA_IDENTIFICATION": True,
    "LOCAL_MODEL": False,
    "GUESS_HANJA": False,
    "DO_CATEGORIZATION": False,
    "DO_TRANSLATION": True,
    "SIMPLIFIED_CHINESE_CONVERSION": False,
    "GENRE": "murim",
    "ERROR_LOG": "error.txt",
    "DICT_API_KEY": ""
}

with open("config.json", "w", encoding="utf-8") as f:
    json.dump(config, f, indent=2, ensure_ascii=False)

print("‚úÖ Created config.json with your API key")

# Now upload files
from google.colab import files
import os

os.makedirs("raws", exist_ok=True)
print("\nüìö Upload your Korean novel .txt files (3 files max for testing)")
uploaded = files.upload()

# Move files
for filename in uploaded.keys():
    os.rename(filename, f"raws/{filename}")
    print(f"‚úì Uploaded: {filename}")

print(f"\n‚úÖ Ready! Uploaded {len(uploaded)} files")

# Run the tool
print("\nüöÄ Starting glossary creation...")
!python cli.py --config config.json

# Check results
print("\nüìä Checking results...")
if os.path.exists("glossary.xlsx"):
    print("‚úÖ Success! glossary.xlsx created")
    files.download("glossary.xlsx")
    files.download("glossary_master.xlsx")
else:
    print("‚ùå No glossary.xlsx created")
    if os.path.exists("error.txt"):
        print("üìÑ Error log:")
        with open("error.txt", "r") as f:
            print(f.read())
```
